```python
#import libraries
import pandas as pd
import numpy as np

#import warning
import warnings
warnings.filterwarnings("ignore")
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
movies = movies.merge(credits , on="title")
```

```python
movies.head()
```

```python
#only importants columns get and make dataframe :

movies = movies[["movie_id" , "title" , "overview","genres" , "keywords" , "cast" , "crew"]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#Null Values Are Droped.
movies.isnull().sum()
```

```python
#check duplicate rows
movies.duplicated().sum()
```

```python
import ast #this library is used to convert string into list format

def convert(data):
    
    update_list = []
    
    for i in ast.literal_eval(data):

        update_list.append(i["name"])
    
    return update_list 
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["genres"].head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies["keywords"].head()
```

```python
def convert_cast(data):
    
    cast_list = []
    counter = 0

    for i in ast.literal_eval(data):

        if counter != 3:

            cast_list.append(i["name"])
            counter += 1
        else:
            break
    return cast_list
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
def fetch_director(data):

    director = []

    for i in ast.literal_eval(data):

        if i["job"] == "Director":

            director.append(i["name"])
            break

    return director
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
#remove spaces because extra spaces are not fitted in ML model
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
#remove spaces
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
#drop columns
drop_columns = ["overview" , "genres" , "keywords" , "cast" , "crew"]
movies.drop(drop_columns , axis=1, inplace=True)
```

```python
movies.head()
```

```python
#join the tags lists and convert into string
movies["tags"] = movies["tags"].apply(lambda x : " ".join(x))
```

```python
movies["tags"].head()
```

```python
#convert into lower text
movies["tags"] = movies["tags"].str.lower()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    text_list = []
    for i in text.split():
        text_list.append(ps.stem(i))
    return " ".join(text_list)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
movies.head()
```

```python
# for Text Vectorize :
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000 , stop_words="english")
cv
```

```python
vectors = cv.fit_transform(movies["tags"]).toarray()
```

```python
vectors
```

```python
#count a similarity for distance of movies
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python
def recommender(movie):
    
    movie_index = movies[movies["title"] == movie].index[0]
    
    distance = similarity[movie_index]

    movie_list = sorted(list(enumerate(distance)),reverse=True , key=lambda x : x[1])[1:6]
    print(f"After You Watch {movie} , So Please Watch This 5 Movies ")
    for i in movie_list:
        
        print(movies.iloc[i[0]].title)
```

```python
#our model is ready 
recommender("Avatar")
```

```python
#dump the model and saved
import pickle

with open("../models/movies_model.pkl" , "wb") as f:
    pickle.dump(movies , f)
```
